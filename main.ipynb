```python
# Initiate model
from ultralytics import YOLOv10

MODEL_PATH = 'yolov10n.pt'
model = YOLOv10(MODEL_PATH)
```

```python
# Train the model
YAML_PATH = '../safety_helmet_dataset/data.yaml'
EPOCHS = 30
IMG_SIZE = 480
BATCH_SIZE = 128

model.train(data=YAML_PATH, epochs=EPOCHS, batch=BATCH_SIZE, imgsz=IMG_SIZE)
```

```python
# Evaluate the model
TRAINED_MODEL_PATH = 'runs/detect/train/weights/best.pt'
model = YOLOv10(TRAINED_MODEL_PATH)

model.val(data=YAML_PATH, imgsz=IMG_SIZE, split='test')
```

```python
# Predict with an image
source = '../safety_helmet_dataset/test/images/helmet-269-_jpg.rf.2b8e58c7477defa6f098ffab8bb34c33.jpg'
model.predict(source=source, save=True)
```
